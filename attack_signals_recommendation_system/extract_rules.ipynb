```python
import pickle
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics, datasets, ensemble
```

```python
kdd99_40raw = pd.read_csv('kdd99_raw40.csv')
print(kdd99_40raw.shape)
kdd99_40raw.head()
```

```python
feature_names = kdd99_40raw.columns[0:-2]
print(len(feature_names))
```

```python
clf = pickle.load(open('random_forest_best.sav', 'rb'))
```

```python
estimator = clf.estimators_[0]
```

```python
def get_code(tree, feature_names, tabdepth=0):
    left      = tree.tree_.children_left
    right     = tree.tree_.children_right
    threshold = tree.tree_.threshold
    features  = [feature_names[i] for i in tree.tree_.feature]
    value = tree.tree_.value

    def recurse(left, right, threshold, features, node, tabdepth=0):
            if (threshold[node] != -2):
                    print ('\t' * tabdepth + "if ( " + features[node] + " <= " + str(threshold[node]) + " ) {")
                    if left[node] != -1:
                            recurse (left, right, threshold, features,left[node], tabdepth+1)
                    print ('\t' * tabdepth + "} else {")
                    if right[node] != -1:
                            recurse (left, right, threshold, features,right[node], tabdepth+1)
                    print ('\t' * tabdepth + "}")
            else:
                    print ('\t' * tabdepth + "return " + str(value[node]))

    recurse(left, right, threshold, features, 0)
```

```python
get_code(estimator, feature_names, tabdepth=2)
```
